```python
import pandas as pd
import numpy as np
from pycaret.classification import *
```

```python
df = pd.read_csv("datasets/df_merged.tar.gz")
print("df shape:", df.shape)
```

```python
df[df["pid"] == 1]
```

```python
def assign_grps(df, ngrps=5):
    
    pid_grp = {}
    df_ = df.copy()

    i = 0
    for pid in df["pid"].unique():
        pid_grp[pid] = i
        i = (i+1) % ngrps

    df_["grp"] = df_["pid"].apply(lambda x: pid_grp[x])
    return df_
```

```python
df = assign_grps(df, 5)
df
```

```python
df["sleep"] = df["sleep_phase"] > 0
```

```python
hr_keys = [k for k in df.keys() if k.startswith("hr")]
act_keys = [k for k in df.keys() if k.startswith("act")]
act_keys
```

```python
s = setup(df, 
          target='sleep',
          ignore_features=["pid", "sleep_phase", "time", "grp"],
          numeric_features=act_keys + hr_keys,
          fold_strategy="groupkfold", fold=5, fold_groups="grp", 
          )
```

```python
compare_models()
```

```python
model = create_model("et")
```

```python
model = tune_model(model, n_iter=3, choose_better=True)
```

```python
evaluate_model(model)
```
